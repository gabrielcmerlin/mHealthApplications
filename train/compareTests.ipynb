# Funções Utilizadas

```python
def print_absolute_mean(data):
  sum = 0

  for i in range(5):
    media_iteracao = np.mean(data[i].tolist())
    print(media_iteracao, end=" ")
    sum += media_iteracao

  print("---> ", sum / 5)
```
